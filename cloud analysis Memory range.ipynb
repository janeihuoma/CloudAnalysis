```python
def get_cdf(data):
    """Returns the CDF of the given data.
    
       Args:
           data: A list of numerical values.
           
       Returns:
           An pair of lists (x, y) for plotting the CDF.
    """
    sorted_data = sorted(data)
    p = 100. * np.arange(len(sorted_data)) / (len(sorted_data) - 1)
    return sorted_data, p
```

```python
#Plot graph of tasks submitted against CDF
x, y = get_cdf(user_task_count['job_name'])
plt.plot(x, y, label='Tasks submitted',  color='green', linestyle='-')
plt.grid(alpha=.3, linestyle='--')
plt.xscale('log')
plt.xlim(10 ** -1, 10 ** 6)
plt.ylim(0, 100)
plt.legend(loc='lower right')
plt.xlabel('Num of tasks submitted per User')
plt.ylabel('CDF')
plt.show()
```

```python
from sklearn.preprocessing import QuantileTransformer
```

```python
qt = QuantileTransformer()
```

```python
x_train = qt.fit_transform(x_train)
```

```python
x_test = qt.fit_transform(x_test)
```

```python
y_train = qt.fit_transform(y_train)
y_test = qt.fit_transform(y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
```

```python
#RF range without cross validation
```

```python
k_range = list(range(1, 5))
msearr = []
rmsearr = []


for k in k_range:
    RFRegressor = RandomForestRegressor(n_estimators=k, random_state=0)
    RFRegressor.fit(x_train, y_train)
    prediction = RFRegressor.predict(x_test)
    mse=(mean_squared_error(y_test, prediction))
    msearr.append(mse)
    rmsearr.append(np.sqrt(mse))
    
print("Mean MSE:", msearr)
print()
print("Mean RMSE:", rmsearr)
```

```python
plt.plot(k_range, msearr)
plt.xlabel('Value of estimators for RF')
plt.ylabel('MSE')
plt.show()
```

```python
plt.plot(k_range, rmsearr)
plt.xlabel('Value of estimators for RF')
plt.ylabel('RMSE')
plt.show()
```
